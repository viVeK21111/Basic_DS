```python
# 5) Top five run scorers

from numpy import *
rdf=df2['runs'].values # converting them to array
t=DataFrame()
for i in range(5):
  a=df2[df2['runs']==rdf.max()]
  t=t.append(a)
  index=argwhere(rdf==rdf.max())
  rdf=delete(rdf,index)
t
```

```python
# 6) Graph of top five scorers

import matplotlib.pyplot as plt
rns=t['runs'].values
ply=t['Player'].values
x=Series(ply)
y=Series(rns)
plt.xlabel=('Player')
plt.ylabel=('Runs scored')
plt.title('Top five run scorers')
plt.bar(x,y)
plt.show()
```

```python
# 7) wicket takers
from numpy import *
rdf=df1['wickets'].values # converting them to array
t1=DataFrame()
for i in range(5):
  a=df1[df1['wickets']==rdf.max()]
  t1=t1.append(a)
  index=argwhere(rdf==rdf.max())
  rdf=delete(rdf,index)
t1
```

```python
# 8) bar graph for wicket takers

import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)
rns=t1['wickets'].values
ply=t1['Player'].values
x=Series(ply)
y=Series(rns)
plt.bar(x,y)
plt.xticks(rotation=30)
plt.title('Top five wicket takers')
plt.xlabel('player') 
plt.ylabel('wickets taken')
plt.show()
```

```python
# 16) highest wickets in an innings
from fractions import Fraction
from numpy import *
best=df1['best'].values # numpy array
l=[]
for i in range(len(best)):
  l.append(Fraction(best[i]))
ln=[]
for i in range(len(best)):
  ln.append(l[i].numerator)
ld=[]
for i in range(len(best)):
  ld.append(l[i].denominator)
d2=array(ld)
d1=array(ln)
index=int(argwhere(d1==d1.max()))
str=('{0}/{1}'.format(d1.max(),d2[index]))
# now modify the existing dataframe and take that to a new dataframe
# change the 'best' in to the only numerators in df1
df1[df1['best']==str]
```

```python
# 18) # percetage of runs and wickets 
# creating data frames for both india and england
edf=df2[0:18] # england
idf=df2[18:]  # india
```

```python
# Run percentage of england
from pandas import *
z=DataFrame()
for i in range(18):
  a={'player':edf['Player'][i],'Run %': '%.2f' %(edf['runs'][i]/edf['innings'][i])} 
  z=z.append(a,ignore_index=True)
z                                                            
```

```python
# Run percentage of india
z1=DataFrame()
for i in range(18,31):
  a={'player':idf['Player'][i],'Run %': '%.2f' %(idf['runs'][i]/idf['innings'][i])} 
  z1=z1.append(a,ignore_index=True)
z1
```

```python
# 19) pie chart of england run percentage
from numpy import *
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
rp=z['Run %'].values
pp=z['player'].values
plt.pie(rp,autopct='%.2f%%',labels=pp)
plt.title('pie chart of england run percentage')
plt.show()
```

```python
# india
from numpy import *
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
rp=z1['Run %'].values
pp=z1['player'].values
plt.pie(rp,autopct='%.2f%%',labels=pp)
plt.title('pie chart of india run percentage')
plt.show()
```
